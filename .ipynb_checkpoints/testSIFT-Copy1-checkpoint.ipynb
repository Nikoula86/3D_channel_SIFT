```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from tifffile import imread, imsave
import os
from glob import glob
from tqdm import tqdm
```

```python
path = '../../../Images/20181109_kdrl_stains/fish02/'
os.path.exists(path)
```

```python
flist = glob(path+'*_ill00.raw')
flist.sort()
print(flist)
```

```python
imgs = np.stack( [ np.reshape(np.frombuffer(open(i,'rb').read(),np.uint16)[:2048*2048*200],(200,2048,2048)) for i in tqdm(flist) ] )
tmp = np.max(imgs,1)
mpis = np.zeros((2048,2048,4))
for i, img in enumerate(tmp):
    _min = np.min(img)
    _max = np.percentile(img,99.7)
    new=np.clip(img,_min,_max).astype(np.float32)
    mpis[...,i] = ((2**8-1)*((new-np.min(new))/(np.max(new)-np.min(new)))).astype(np.uint8)

print('Data input shape: ', imgs.shape, mpis.shape)
```

```python
mpis=mpis.astype(np.uint8)
fig,ax=plt.subplots(figsize=(12,12),nrows=2,ncols=2)
ax=ax.flatten()
ch = ['488nm','561nm','647nm','800nm']
for i in range(mpis.shape[-1]):
    ax[i].imshow(mpis[...,i])
    ax[i].set_xlabel(ch[i])
print(mpis.shape,mpis.dtype)

from tifffile import imsave
#for i in range(4):
#    imsave('mpisOrig_%d.tif'%i,mpis[...,i])
```

```python
'''
source:
https://ianlondon.github.io/blog/how-to-sift-opencv/
'''

def gen_sift_features(gray_img):
    sift = cv2.xfeatures2d.SIFT_create()
    # kp is the keypoints
    #
    # desc is the SIFT descriptors, they're 128-dimensional vectors
    # that we can use for our final features
    kp, desc = sift.detectAndCompute(gray_img, None)
    return kp, desc

def show_sift_features(gray_img, color_img, kp, ax):
    return ax.imshow(cv2.drawKeypoints(gray_img, kp, color_img.copy(),flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS))
```

```python
fig,ax=plt.subplots(figsize=(12,12),nrows=2,ncols=2)
ax=ax.flatten()
ch = ['488nm','561nm','647nm','800nm']

kp = [[],[],[],[]]
desc = [[],[],[],[]]
for i in range(mpis.shape[-1]):
    ax[i].set_xlabel(ch[i])
    kp[i], desc[i] = gen_sift_features(mpis[...,i])
    show_sift_features(mpis[...,i], mpis[...,i], kp[i], ax[i]);
```

```python
fig,ax=plt.subplots(figsize=(12,18),nrows=3,ncols=1)
ax=ax.flatten()
ch = ['488nm','561nm','647nm','800nm']

bf = cv2.BFMatcher(cv2.NORM_L2,crossCheck=True)

matches = [[],[],[]]
for i in range(3):
    m=bf.match(desc[0],desc[i+1])
    m=sorted(m,key=lambda x: x.distance)
    matches[i]=m[:25]
    img=cv2.drawMatches(mpis[...,0],kp[0],mpis[...,i+1],kp[i+1],matches[i],mpis[...,i+1].copy(),flags=2)
    ax[i].imshow(img)
```

```python
'''
source:
https://www.learnopencv.com/image-alignment-feature-based-using-opencv-c-python/
'''

h = [None,None,None]
mask = [None,None,None]

for i in range(3):
    print(i,len(matches[i]),len(kp[i+1]))
    m = matches[i]
    p1 = np.zeros((len(m),2),dtype=np.float32)
    p2 = np.zeros((len(m),2),dtype=np.float32)
    print(p1.shape,p2.shape)
    for j, match in enumerate(m):
#        print(j)
        p1[j,:]=kp[0][match.queryIdx].pt
        p2[j,:]=kp[i+1][match.trainIdx].pt
    h[i],mask[i]=cv2.findHomography(p2,p1,cv2.RANSAC)
```

```python
mpisReg = 0*mpis
mpisReg[...,0]=mpis[...,0]
height,width=mpis[...,0].shape
for i in range(3):
    mpisReg[...,i+1]=cv2.warpPerspective(mpis[...,i+1],h[i],(width,height))
```

```python
fig,ax=plt.subplots(figsize=(12,12),nrows=2,ncols=2)
ax=ax.flatten()
ch = ['488nm','561nm','647nm','800nm']
for i in range(mpis.shape[-1]):
    ax[i].imshow(mpisReg[...,i])
    ax[i].set_xlabel(ch[i])
print(mpis.shape,mpis.dtype)
```

```python
from tifffile import imsave
for i in range(4):
    imsave('mpisReg_%d.tif'%i,mpisReg[...,i])
```

```python
imgsXYreg = 0*imgs
print(imgsXYreg.shape, imgsXYreg.dtype)
```

```python
print('registering in XY...')
h = [np.eye(3),h[0],h[1],h[2]]
for i in tqdm( range(imgsXYreg.shape[1]) ):
    for j in range(4):
        imgsXYreg[j,i] = cv2.warpPerspective(imgs[j,i],h[j],(width,height))
```

```python
for i in tqdm(range(4)):
    imsave('ch%02d_ill00_XYreg.tif'%i,imgsXYreg[i])
```
